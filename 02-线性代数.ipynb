# 6. 线性代数
## 6.1 标量
    标量由只有一个元素的张量表示。

```python
import torch
from d2l.tensorflow import reshape

x = torch.tensor([3.0])
y = torch.tensor([2.0])

x + y, x * y, x / y, x ** y
```

## 6.2 向量
### 6.2.1 创建向量
    可以将向量视为标量值组成的列表。

```python
x = torch.arange(4)
x
```

### 6.2.2 访问向量元素
    通过张量的索引来访问任一元素。

```python
import torch
x = torch.arange(4)
print(x[3])  # 索引从0开始
```

### 6.2.3 访问向量长度
    访问张量的长度。

```python
import torch
x = torch.arange(4)
len(x)
```

### 6.2.4 访问向量维度
    只有一个轴的张量，形状只有一个元素。

```python
import torch
x = torch.arange(4)
x.shape
```

## 6.3 矩阵
### 6.3.1 创建矩阵
    通过指定两个分量m和n来创建一个形状为m×n的矩阵。

```python
import torch
A = torch.arange(20).reshape(5, 4)
A
```

### 6.3.2 矩阵转置
    矩阵的转置。

```python
A = torch.arange(20).reshape(5, 4)
A.T
```

### 6.3.3 对称矩阵
对称矩阵（symmetric matrix）A 等于其转置：$A = A^{T}$
